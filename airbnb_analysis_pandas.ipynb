```python
plt.figure(figsize=(10,5))
sns.histplot(df['price'],bins=50,kde=True)
plt.title("Price Distribuition of Airbnb Listings")
plt.xlabel('Price (in USD)')
plt.ylabel("Frequency")
plt.show()
```

### Price Distribution:
An attempt was made to display the full distribution of Airbnb listing prices. However, it’s clear that most prices are concentrated around **\$0**, with the x-axis stretching all the way up to **\$100,000**.  
This happens because a few extremely high-priced listings (outliers) distort the distribution, making it harder to accurately represent the prices of most listings.  
As a result, the majority of listings, which are priced much lower, appear squeezed into a small section of the plot, making it challenging to identify any meaningful trends


```python
bins = [0, 200, 400, 600, 1000, 10000,200000]

bin_labels = ['0-200', '200-400', '400-600', '600-1k', '1k-10k','10k-200k']

df['price_bin'] = pd.cut(df['price'], bins=bins, labels=bin_labels, include_lowest=True)

plt.figure(figsize=(10,6))
sns.countplot(x='price_bin', data=df, palette='muted')
plt.title('Distribution of Airbnb Listings by Price Range')
plt.xlabel('Price Range (in USD)')
plt.ylabel('Number of Listings')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Price Distribution:
It’s clear that most Airbnb listings are priced between **\$0** and **\$200**, with a significant number falling in the **\$200**-**\$400** range.  
There are fewer listings in the **\$400**-**\$600** price range, and the number of listings drops sharply beyond $600.  

Only a small fraction of listings are priced above **\$1,000**, and extremely high-priced listings (e.g., above **\$10,000**) make up a very small part of the data.  
These outliers can skew visualizations, making it harder to focus on the majority of listings that are priced more typically.  

To enhance clarity, outliers will be addressed by applying a price cap and filtering the data, allowing for a more focused analysis of the majority of Airbnb listings


```python
filtered_df = df[df['price'] <= 1000]
```

```python
plt.figure(figsize=(10,6))
sns.histplot(filtered_df['price'], bins=50, kde=True)
plt.title('Price Distribution of Airbnb Listings (Price <= $1000)')
plt.xlabel('Price (in USD)')
plt.ylabel('Number of Listings')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Adjusted Price Distribution:
In this second plot, the dataset was limited to include only listings priced at \$1,000 or below, which covers the vast majority of listings.  
As a result, the distribution is now much clearer:

Most listings are priced between **\$50** and **\$300**, with a peak around **\$100** per night.  
The distribution is right-skewed, meaning that while there are many affordable listings, a small number of higher-priced listings still appear.  
This plot provides a much clearer picture of the typical prices for Airbnb listings in New York City, free from the distortion caused by extreme outliers.


```python
bins = [0, 100, 200, 400, 800, 1000]
bin_labels = ['0-100', '100-200', '200-400', '400-800', '800-1000']

filtered_df['price_bin'] = pd.cut(filtered_df['price'], bins=bins, labels=bin_labels, include_lowest=True)

plt.figure(figsize=(10,8))
sns.countplot(x='price_bin', hue='room_type', data=filtered_df, palette='muted')

plt.title('Price Distribution by Room Type (Binned Price Ranges)', fontsize=16)
plt.xlabel('Price Range (in USD)', fontsize=14)
plt.ylabel('Number of Listings', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(title='Room Type', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

### Price Concentration:
Most Airbnb listings are concentrated in the lower price ranges, especially for private rooms and entire homes/apartments.  
This supports the idea that Airbnb is primarily a platform for more affordable, short-term stays.  
As prices rise, the number of listings drops sharply, with only a small number of high-end listings exceeding $1,000 USD.

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='room_type', y='availability_365', data=filtered_df, palette='muted')

plt.title('Availability Distribution by Room Type', fontsize=16)
plt.xlabel('Room Type', fontsize=14)
plt.ylabel('Number of Available Days (availability_365)', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Availability by Property Type:

- **Hotel Rooms Have the Highest Availability**  
The median availability for hotel rooms is close to 365 days per year, indicating that they are almost always available.  
The interquartile range (IQR) is small, meaning most hotels maintain high availability.  
There are a few outliers with much lower availability, possibly due to seasonal hotels.

- **Private and Shared Rooms Show a Wide Spread**  
Private rooms show a broad range of availability, with some listings offering close to 0 days, while others are available nearly all year round.  
Shared rooms follow a similar trend, suggesting that some hosts rent them full-time, while others offer them seasonally.

- **Entire Apartments Have a Balanced Distribution**  
The median availability for entire apartments is around 200 days, indicating that many are available only part of the year.  
The IQR is quite large, showing that there are varied usage patterns—some are short-term rentals, while others operate year-round.

### Key Takeaways:
- Hotels tend to have the most consistent availability.
- Private and shared rooms have a wide variation in availability, likely due to casual or part-time renting.
- Entire apartments are used for both long-term and seasonal rentals, giving them a more balanced availability.


```python
bins = [0, 90, 180, 270, 365]
labels = ['0-90 days', '91-180 days', '181-270 days', '271-365 days']

filtered_df['availability_category'] = pd.cut(filtered_df['availability_365'], bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(10,6))
sns.boxplot(x='availability_category', y='price', data=filtered_df, palette='muted')

plt.title('Rental Price Distribution Based on Availability', fontsize=16)
plt.xlabel('Availability Category (Days per Year)', fontsize=14)
plt.ylabel('Rental Price (USD)', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Rental Price Distribution by Availability:

This boxplot visualizes the distribution of rental prices across different availability categories. The dataset is divided into four groups based on the number of days a listing is available per year:  
- 0-90 days  
- 91-180 days  
- 181-270 days  
- 271-365 days  

The y-axis represents the rental price in USD.

### Key Observations:

- **No Strong Relationship Between Availability and Price**  
The median rental price remains relatively stable across all availability categories, suggesting that a listing’s availability does not significantly affect its price.

- **Luxury Listings in Every Category**  
The presence of several high-priced outliers indicates that luxury properties exist across all availability categories. These outliers, with prices reaching several thousand dollars per night, skew the overall price distribution.

- **Most Listings Fall in the Lower Price Range**  
Most listings are priced toward the lower end of the scale, as seen in the compact boxplots. This suggests that, despite the presence of high-end properties, the majority of Airbnb rentals are more affordable.

This plot highlights that rental prices remain consistent across different availability levels, with a few expensive listings pushing the upper limits. Availability doesn’t seem to have a clear impact on rental prices, as median prices stay similar across all categories. However, high-priced outliers show that luxury properties can be available both year-round and for shorter periods.
